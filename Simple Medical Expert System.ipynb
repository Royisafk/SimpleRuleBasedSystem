```python
#Building a Simple Medical Expert System Using Knowledge Representation
class MedicalExpertSystem:

    def __init__(self):
        self.knowledge_base = {
            "Malaria": ["Fever", "Headache", "Fatigue"],
            "Pneumonia": ["Cough", "Chest Pain", "Fatigue"],
            "Flu": ["Sneezing", "Runny Nose", "Sore Throat"],
            "Food Poisoning": ["Vomiting", "Diarrhea", "Fatigue"]
        }
    def diagnose(self, symptoms):
        diseases = [] 
        for disease, disease_symptoms in self.knowledge_base.items():
            if all(symptom in symptoms for symptom in disease_symptoms):
                diseases.append(disease)

        return diseases
system = MedicalExpertSystem()
user_input = input("Enter symptoms separated by commas: ")
symptoms = [s.strip().title() for s in user_input.split(",")]
result = system.diagnose(symptoms)
if result:
    print("Possible Illness:")
    for disease in result:
        print("-", disease)
else:
    print("No matching illness found.")
```
